```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
results = pd.read_csv('baselines/dehnn_50_3_16_True_False_2025-01-25_16-52-26_eigenvectors.csv')
results
```

```python
plt.plot(results['Epoch'], results['Node_Valid_Loss'], label='Validation Loss')
plt.plot(results['Epoch'], results['Node_Train_Loss'], label='Training Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Node Loss')
plt.legend()
plt.show()
```

```python
plt.plot(results['Epoch'], results['Net_Valid_Loss'], label='Validation Loss')
plt.plot(results['Epoch'], results['Net_Train_Loss'], label='Training Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Net Loss')
plt.legend()
plt.show()
```
