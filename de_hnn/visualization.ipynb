```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
results = pd.read_csv('baselines/dehnn_500_2_16_True_False_2025-02-02_23-26-08_classify.csv')
results
```

```python
plt.plot(results['Epoch'], results['fsc'])
```

```python
plt.plot(results['Epoch'], results['Node_Valid_Loss'], label='Validation Loss')
plt.plot(results['Epoch'], results['Node_Train_Loss'], label='Training Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Node Loss')
plt.legend()
plt.show()
```

```python
plt.plot(results['Epoch'], results['Net_Valid_Loss'], label='Validation Loss')
plt.plot(results['Epoch'], results['Net_Train_Loss'], label='Training Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Net Loss')
plt.legend()
plt.show()
```
