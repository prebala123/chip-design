```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
baseline = pd.read_csv('../results/baselines/2025-02-05_19-13-38__200_2_16_True_congestion_baseline.csv')
persist = pd.read_csv('../results/baselines/2025-02-06_16-56-58__200_2_16_True_congestion_pd.csv')
evects = pd.read_csv('../results/baselines/2025-02-06_11-59-11__200_2_16_True_congestion_evects.csv')
baseline
```

```python
metrics = np.array([['FscoreNet', 'PrecisionNet', 'RecallNet', 'NetTrain', 'NetValid'],
           ['FscoreNode', 'PrecisionNode', 'RecallNode', 'NodeTrain', 'NodeValid']]).T
fig, ax = plt.subplots(5, 2, figsize=(15, 20))
plt.subplots_adjust(hspace=0.3)
# fig.suptitle('Baseline Results vs No Persistence Diagram')
for i in range(len(metrics)):
    for j in range(len(metrics[i])):
        ax[i, j].plot(baseline['Epoch'], baseline[metrics[i][j]], label='Baseline')
        ax[i, j].plot(persist['Epoch'], persist[metrics[i][j]], label='No PD')
        ax[i, j].plot(evects['Epoch'], evects[metrics[i][j]], label='No Eigenvectors')
        ax[i, j].set_title(metrics[i][j])
        ax[i, j].set(xlabel='Epochs')
        ax[i, j].legend()

fig.savefig("../results/plots/pd.png", dpi=300, bbox_inches='tight')
plt.show()
```
