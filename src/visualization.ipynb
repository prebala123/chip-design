```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
baseline = pd.read_csv('../results/ablation/baseline_500.csv')

fig, ax = plt.subplots(1, 3, figsize=(16, 5))
fig.suptitle('Baseline', fontsize=16)

ax[0].plot(baseline['Epoch'], baseline['Train_RMSE'], label='Train Loss')
ax[0].plot(baseline['Epoch'], baseline['Valid_RMSE'], label='Valid Loss')
ax[0].plot(baseline['Epoch'], baseline['Test_RMSE'], label='Test Loss')
ax[0].legend()
ax[0].set_title('RMSE')
ax[0].set_xlabel('Epochs')

ax[1].plot(baseline['Epoch'], baseline['Train_MAE'], label='Train Loss')
ax[1].plot(baseline['Epoch'], baseline['Valid_MAE'], label='Valid Loss')
ax[1].plot(baseline['Epoch'], baseline['Test_MAE'], label='Test Loss')
ax[1].legend()
ax[1].set_title('MAE')
ax[1].set_xlabel('Epochs')

ax[2].plot(baseline['Epoch'], baseline['Train_Pearson'], label='Train Loss')
ax[2].plot(baseline['Epoch'], baseline['Valid_Pearson'], label='Valid Loss')
ax[2].plot(baseline['Epoch'], baseline['Test_Pearson'], label='Test Loss')
ax[2].legend()
ax[2].set_title('Pearson')
ax[2].set_xlabel('Epochs')

plt.show()
```

```python
comparison_name = 'Eigenvectors'

baseline = pd.read_csv('../results/baselines/2025-02-07_18-24-57__200_2_16_True_congestion_baseline.csv')
evects = pd.read_csv('../results/baselines/2025-02-06_11-59-11__200_2_16_True_congestion_evects.csv')
```

```python
metrics = [['FscoreNet', 'PrecisionNet', 'RecallNet'],
           ['NetTrain', 'NetValid']]
fig, ax = plt.subplots(1, 3, figsize=(20, 5))
fig.suptitle(f'Baseline Results vs No {comparison_name}')

ax[0].plot(baseline['Epoch'], baseline['PrecisionNet'], label='Baseline')
ax[0].plot(evects['Epoch'], evects['PrecisionNet'], label=comparison_name)
ax[0].set_title('Precision')
ax[0].set(xlabel='Epochs')
ax[0].legend()

ax[1].plot(baseline['Epoch'], baseline['RecallNet'], label='Baseline')
ax[1].plot(evects['Epoch'], evects['RecallNet'], label=comparison_name)
ax[1].set_title('Recall')
ax[1].set(xlabel='Epochs')
ax[1].legend()

ax[2].plot(baseline['Epoch'], baseline['FscoreNet'], label='Baseline')
ax[2].plot(evects['Epoch'], evects['FscoreNet'], label=comparison_name)
ax[2].set_title('F_score')
ax[2].set(xlabel='Epochs')
ax[2].legend()

fig.savefig(f"../results/plots/{comparison_name}.png", dpi=300, bbox_inches='tight')
plt.show()
```
